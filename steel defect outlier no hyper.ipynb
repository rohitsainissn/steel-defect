```python
#1 
from scipy import stats

# Calculate Z-scores for each column in X_train and X_cv
z_scores_train = stats.zscore(X_train1)
z_scores_cv = stats.zscore(X_cv1)

# Define a threshold for outlier detection (e.g., 3 standard deviations)
threshold = 3

# Filter out rows with outliers in X_train
X_train0 = X_train1[(z_scores_train < threshold).all(axis=1)]

# Filter out rows with outliers in X_cv
X_cv0= X_cv1[(z_scores_cv < threshold).all(axis=1)]


filtered_indices_train = X_train1.index.difference(X_train0.index)
filtered_indices_cv = X_cv1.index.difference(X_cv0.index)

# Remove the same rows from y_train_all and y_cv_all
y_train1 = y_train1.drop(filtered_indices_train)
y_cv1 = y_cv1.drop(filtered_indices_cv)
X_train1,X_cv1=X_train0,X_cv0
```
